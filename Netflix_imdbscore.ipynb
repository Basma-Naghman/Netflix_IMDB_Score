```python
import nltk
from nltk.corpus import stopwords
import string
string.punctuation
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y =[]
    for i in text:
        if i.isalnum():
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)


    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))


    return " ".join(y)      




```

```python
transform_text("I love youtube lecture on ML")
```

```python
df1['transformed_title'] = df1['title'].apply(transform_text)
```

```python
df1['transformed_description'] = df1['description'].apply(transform_text)
```

```python
df1 = df1.drop(columns=["title","description","imdb_votes"])
```

```python
df1["imdb_votes_scaled"].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
df1['imdb_votes_scaled'] = scaler.fit_transform(df1[['imdb_votes']])
```

```python
df1.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np


# 2. Define transformers
title_vectorizer = CountVectorizer(max_features=2000)
description_vectorizer = CountVectorizer(max_features=2000)

other_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first'), 
        ['type', 'age_certification']),
        ('num', StandardScaler(), 
        ['runtime'])
    ],
    remainder='drop'
)

# 3. Create features
title_features = title_vectorizer.fit_transform(df1['transformed_title'])
description_features = description_vectorizer.fit_transform(df1['transformed_description'])
other_features = other_transformer.fit_transform(df1)

# 4. Verify shapes match
print(f"Title features shape: {title_features.shape}")
print(f"Description features shape: {description_features.shape}")
print(f"Other features shape: {other_features.shape}")

# 5. Combine all features
X_transformed = np.hstack([
    other_features,
    title_features.toarray(),
    description_features.toarray()
])

# 6. Get feature names
cat_features = other_transformer.named_transformers_['cat'].get_feature_names_out(['type', 'age_certification'])
num_features = ['runtime'] 
title_features_names = ["title_"+name for name in title_vectorizer.get_feature_names_out()]
desc_features_names = ["desc_"+name for name in description_vectorizer.get_feature_names_out()]

all_feature_names = np.concatenate([
    cat_features, 
    num_features, 
    title_features_names,
    desc_features_names
])

# 7. Create final DataFrame
transformed_df = pd.DataFrame(X_transformed, columns=all_feature_names)

# 8. Add target variables if needed
transformed_df['imdb_score'] = df1['imdb_score'].values
transformed_df['imdb_votes_scaled'] = df1['imdb_votes_scaled'].values
```
